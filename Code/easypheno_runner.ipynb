```python
# 读取文件（用你自己的路径）
df = pd.read_csv('../Data/x_matrix.csv')

# 定义允许的字符（IUPAC raw 模式）
allowed_iupac = {'A', 'C', 'G', 'T', 'M', 'R', 'W', 'S', 'Y', 'K'}

# 替换所有不属于 allowed_iupac 的字符为 'N'
df.iloc[:, 1:] = df.iloc[:, 1:].applymap(lambda x: x if x in allowed_iupac else 'N')

# 保存为新的 raw 格式文件
df.to_csv('../Data/x_matrix_raw.csv', index=False)
```
